```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Pandas/data/data03.csv')
data
```

```python
data['남자+여자'] = data['남자'] + data['여자']
data
```

```python
data['남자-여자'] = data['남자'] - data['여자']
data
```

```python
data['남자-여자_절댓값'] = data['남자-여자'].abs()
data
```

```python
data['남자/여자'] = data['남자']/data['여자']
data
```

```python
data['여자_10명당_남자수'] =data['남자/여자']+10
data
```
